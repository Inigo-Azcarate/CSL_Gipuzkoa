```python
def save_data(Per_hog, Turismos, Edad, crnt_tur, Tipo_familia, model_result, r_w):
    data = pd.read_csv(f"data/data.csv")

    fila1 = pd.Series({'Per_hog': Per_hog, 'Turismos': Turismos, 'Edad': Edad, 'crnt_tur': crnt_tur,'Tipo_familia': Tipo_familia, 'Scenario': '1', 'Model_result': model_result[0], 'R_W': r_w[0]})
    fila2 = pd.Series({'Per_hog': Per_hog, 'Turismos': Turismos, 'Edad': Edad, 'crnt_tur': crnt_tur,'Tipo_familia': Tipo_familia, 'Scenario': '2', 'Model_result': model_result[1], 'R_W': r_w[1]})
    fila3 = pd.Series({'Per_hog': Per_hog, 'Turismos': Turismos, 'Edad': Edad, 'crnt_tur': crnt_tur,'Tipo_familia': Tipo_familia, 'Scenario': '3', 'Model_result': model_result[2], 'R_W': r_w[2]})
   
    data = pd.concat([data, pd.DataFrame([fila1]), pd.DataFrame([fila2]), pd.DataFrame([fila3])], ignore_index=True)
    data.to_csv(f"data/data.csv", index=False)
```

```python
save_data(1,1,1,1,1,['Coche', 'TP', 'TP'], ['Right', 'Right', 'Wrong'])
```
